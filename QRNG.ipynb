```python
from qiskit import QuantumCircuit, transpile
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, plot_state_city



number_of_qbits = 10
iterations = 3

qc = QuantumCircuit(number_of_qbits) # The Quantum circuit

# Apply Hadamard gate to each qubit to create superposition
for i in range(number_of_qbits):
    qc.h(i)

state = Statevector(qc)

qc.measure_all()
qc.draw('mpl')
```

```python
simulator = AerSimulator()
circ = transpile(qc, simulator)

for i in range(iterations):
    # Run and get counts

    result = simulator.run(circ).result()
    counts = result.get_counts(circ)

    # Get the single key (bitstring) from counts
    quantum_bin = list(counts.keys())[0]

    # Convert to integer
    quantum_num = int(quantum_bin, 2)

    # Convert to various formats
    quantum_dec = quantum_num

    print(f"Quantum RNG Results for iteration {i+1}:")
    print("Binary:      ", quantum_bin)
    print("Decimal:     ", quantum_dec)
```
